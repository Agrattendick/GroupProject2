```python
query ="""SELECT State_Name, Murders, murder.state, population, murder_rate, COUNT(formatted_address)
FROM (murder 
    LEft JOIN bj_loc ON murder.state=bj_loc.state)
    GROUP BY state;"""
Murder_Ufo= pd.read_sql(query, conn)
```

```python
Murder_Ufo.rename(columns={'COUNT(formatted_address)': 'store_count'}, inplace=True)
```

```python
query_ufo = """select state, count(id) from us_ufo_csv
group by state;"""
data_ufo = pd.read_sql(query_ufo, conn)
```

```python
data_ufo
```

```python
data_ufo.rename(columns={ 'count(id)': 'ufo_sightings'}, inplace=True)
```

```python
data_ufo
```

```python
Murder_Ufo["ufo_sightings"] = data_ufo["ufo_sightings"]
Murder_Ufo.head()
```

```python
ufo_rate = []
store_rate = []
for i,r in Murder_Ufo.iterrows():
    ufo_rate.append((r['ufo_sightings']/ r['population']*100000))
    store_rate.append((r['store_count']/ r['population']*100000))
```

```python
Murder_Ufo["ufo_rate"] = ufo_rate
```

```python
Murder_Ufo["store_rate"] = store_rate
Murder_Ufo
```

```python
Murder_Ufo.to_sql('state_data', con=conn, if_exists="replace")
```
